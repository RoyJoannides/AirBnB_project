                                                   Cleaning and Wrangling of Dataset

```python
df.head()
```

```python
# Step 1: Drop 'name' column
df_cleaned = df.drop(['name'], axis=1) # The name of the apartment if not relevant for the analysis. 
# Additionally, there is a corresponding ID column which serves the same purpose of identifying each apartment. 
#In the latter, it has more uniques values (48895) compared to (47905) of the name column
```

```python
# Step 2: Drop 'host_name' column
df_cleaned = df_cleaned.drop(['host_name'], axis=1) # The name of the host is not relevant for the analysis. 
# Additionally, there is a corresponding ID column which serves the same purpose of identifying each apartment. 
#In the latter, it has more uniques values (37457) compared to (11452) of the host_name column.
# In geo-ethics, it is considered best practice to anonymize names of subjects for privacy and safety reasons
```

```python
# Step 3: Drop 'last_review' column # The last_review column is not relevant for the analysis. 
df_cleaned = df_cleaned.drop(['last_review'], axis=1)
```

```python
# Step 4: Rename 'id' column to 'address_id' # "id" is not very descriptive. However, "address_id" is more intuitive for the reader
df_cleaned = df_cleaned.rename(columns={'id': 'address_id'})
```

```python
# Step 5: Rename 'availability_365' column to 'yearly_availability_365' #"availability_365" is not very descriptive. However, "yearly_availability_365" is more intuitive for the reader
df_cleaned = df_cleaned.rename(columns={'availability_365': 'yearly_availability_365'})
```

```python
# Step 6: Fill null/empty values of 'reviews_per_month' column with mean of the column 
reviews_per_month_mean = df_cleaned['reviews_per_month'].mean()
df_cleaned['reviews_per_month'] = df_cleaned['reviews_per_month'].fillna(reviews_per_month_mean)
# Considering the nature of this column,this aggregation method provides a representative and unbiased estimate, 
# minimizing the impact of missing values on the distribution of this columns data which will be used in the analysis.
```

Business case 1 (For planning authority/ minicipality of NYC): Identify and promote listings in low-performing geographic clusters, focusing marketing efforts and optimizing pricing strategies in these areas to decentralize population growth.

                              Visualizing the number of host listings vrs the popularity index: Hypothesis 3

```python
# Scatter plot to visualize the relationship
plt.figure(figsize=(12, 6))
sns.scatterplot(x='calculated_host_listings_count', y='number_of_reviews', data=df_cleaned, alpha=0.5)
plt.title('Scatter Plot: Number of Host Listings vs Total Number of Reviews')
plt.xlabel('Number of Host Listings')
plt.ylabel('Total Number of Reviews')
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = df_cleaned['calculated_host_listings_count'].corr(df_cleaned['number_of_reviews'])
print(f"Correlation Coefficient: {correlation_coefficient}")
```

The correlation coefficient of approximately -0.072 indicates a very weak negative correlation between the number of host listings (calculated_host_listings_count) and the total number of reviews (number_of_reviews).
This means that, based on the data, there is no strong linear relationship between the number of host listings and the total number of reviews. Hosts with more listings do not necessarily receive more reviews, or vice versa.
